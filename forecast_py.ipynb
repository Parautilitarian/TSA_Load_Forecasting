## Load & Processing

### Demand Load

```python
import pandas as pd

load = pd.read_excel("load.xlsx")
load.head()
```

```python
load["meter_id"].value_counts()
```

```python
# df = your dataframe
load = load.drop(columns=["meter_id"])

# reshape from wide to long
load_long = load.melt(id_vars="date", var_name="hour", value_name="demand")

# extract hour number from h1...h24
load_long["hour"] = load_long["hour"].str.extract("(\d+)").astype(int)

# convert date column
load_long["date"] = pd.to_datetime(load_long["date"])

# aggregate hourly values to daily mean
daily_load = load_long.groupby("date")["demand"].mean().reset_index(name="daily_load")

daily_load.head()
```

### Temperature

```python
temp = pd.read_excel("temperature.xlsx")
temp.head()
```

```python
# create datetime column
temp["datetime"] = pd.to_datetime(temp["date"]) + pd.to_timedelta(
    temp["hr"] - 1, unit="h"
)

# drop original columns
temp_final = temp.drop(columns=["date", "hr"])

# move datetime to first column
cols = ["datetime"] + [c for c in temp_final.columns if c != "datetime"]
temp_final = temp_final[cols].sort_values("datetime").reset_index(drop=True)
```

```python
temp_final.tail()
```

### Relative Humidity

```python
rh = pd.read_excel("relative_humidity.xlsx")
rh.head()
```

```python
# create datetime column
rh["datetime"] = pd.to_datetime(rh["date"]) + pd.to_timedelta(rh["hr"] - 1, unit="h")

# drop original columns
rh_final = rh.drop(columns=["date", "hr"])

# move datetime to first column
cols = ["datetime"] + [c for c in rh_final.columns if c != "datetime"]
rh_final = rh_final[cols].sort_values("datetime").reset_index(drop=True)
```

```python
df = (
    load_final.merge(temp_final, on="datetime")
    .merge(temp_final, on="datetime")
    .merge(rh_final, on="datetime")
)
df.columns
```

### Feature Engineering

```python
import numpy as np


# refactorize
def feature_engineering(df0):
    df = df0.copy()

    df["date"] = pd.to_datetime(df["date"])

    # Date features
    df["dayofweek"] = df["date"].dt.dayofweek
    df["month"] = df["date"].dt.month
    df["dayofyear"] = df["date"].dt.dayofyear
    df["weekend"] = (df["dayofweek"] >= 5).astype(int)

    # Cyclic encoding
    df["dow_sin"] = np.sin(2 * np.pi * df["dayofweek"] / 7)
    df["dow_cos"] = np.cos(2 * np.pi * df["dayofweek"] / 7)
    df["doy_sin"] = np.sin(2 * np.pi * df["dayofyear"] / 365)
    df["doy_cos"] = np.cos(2 * np.pi * df["dayofyear"] / 365)

    # Lag features (daily)
    df["day_lag_1"] = df["daily_load"].shift(1)
    df["week_lag"] = df["daily_load"].shift(7)
    df["year_lag"] = df["daily_load"].shift(365)

    # Rolling statistics (daily windows)
    df["rollmean_7"] = df["daily_load"].shift(1).rolling(7).mean()
    df["rollmean_30"] = df["daily_load"].shift(1).rolling(30).mean()
    df["rollstd_7"] = df["daily_load"].shift(1).rolling(7).std()

    # Fourier terms
    df["t"] = np.arange(len(df))
    K = 3
    for k in range(1, K + 1):
        df[f"weekly_sin_{k}"] = np.sin(2 * np.pi * k * df["t"] / 7)
        df[f"weekly_cos_{k}"] = np.cos(2 * np.pi * k * df["t"] / 7)
        df[f"yearly_sin_{k}"] = np.sin(2 * np.pi * k * df["t"] / 365)
        df[f"yearly_cos_{k}"] = np.cos(2 * np.pi * k * df["t"] / 365)

    return df
```

```python
load_fe = feature_engineering(daily_load)
load_fe = load_fe.dropna()

print(f"The data ranges from {load_fe['date'].min()} to {load_fe['date'].max()}")
```

### Dataset Splits

```python
# split train/test and separate features/target
def split_train_test(df):
    df = df.sort_values("date")

    # test set = last year
    test_start = df["date"].max() - pd.DateOffset(years=1)

    train = df[df["date"] < test_start]
    test = df[df["date"] >= test_start]

    X_train = train.drop(columns=["daily_load", "date"])
    y_train = train["daily_load"]

    X_test = test.drop(columns=["daily_load", "date"])
    y_test = test["daily_load"]

    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = split_train_test(load_fe)
```

## Gradient Boost Tree Model

```python
from xgboost import XGBRegressor
from sklearn.model_selection import TimeSeriesSplit, RandomizedSearchCV
from sklearn.metrics import mean_absolute_percentage_error

tscv = TimeSeriesSplit(n_splits=5)

xgb = XGBRegressor(objective="reg:squarederror", random_state=42, n_jobs=-1)
```

```python
# hyperparameter grid for tuning
param_grid = {
    "n_estimators": [300, 500, 800],
    "max_depth": [4, 6, 8],
    "learning_rate": [0.01, 0.05, 0.1],
    "subsample": [0.7, 0.8, 1.0],
    "colsample_bytree": [0.7, 0.8, 1.0],
}

search = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=param_grid,
    n_iter=20,
    cv=tscv,
    scoring="neg_mean_absolute_percentage_error",
    n_jobs=-1,
    verbose=1,
)

search.fit(X_train, y_train)
```

```python
best_model_xgb = search.best_estimator_
```

### Testing

```python
from sklearn.metrics import mean_absolute_percentage_error

y_pred = best_model_xgb.predict(X_test)

mape = mean_absolute_percentage_error(y_test, y_pred)

print(f"MAPE (%): {mape * 100:.3f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(y_test.index, y_test, label="Actual")
plt.plot(y_test.index, y_pred, label="Predicted")
plt.legend()
plt.title("Actual vs Predicted")
plt.show()
```

```python
def export_submission(forecast_values, filename):
    # extract forecast values
    fc_values = pd.Series(forecast_values).astype(float).values

    # create July 2011 dates
    dates = pd.date_range(start="2011-07-01", end="2011-07-31", freq="D")

    # keep only first 31 forecast values
    df = pd.DataFrame({"date": dates, "load": fc_values[: len(dates)]})

    # write csv
    df.to_csv(filename, index=False)
```
