## Load & Processing

### Demand Load

```python
import pandas as pd

load = pd.read_excel("load.xlsx")
load.head()
```

```python
load["meter_id"].value_counts()
```

```python
# df = your dataframe
load = load.drop(columns=["meter_id"])

# reshape from wide to long
load_long = load.melt(id_vars="date", var_name="hour", value_name="demand")

# extract hour number from h1...h24
load_long["hour"] = load_long["hour"].str.extract("(\d+)").astype(int) - 1

# create datetime column
load_long["datetime"] = pd.to_datetime(load_long["date"]) + pd.to_timedelta(
    load_long["hour"], unit="h"
)

# final structure
load_final = (
    load_long[["datetime", "demand"]].sort_values("datetime").reset_index(drop=True)
)
```

```python
load_final.tail()
```

### Temperature

```python
temp = pd.read_excel("temperature.xlsx")
temp.head()
```

```python
# create datetime column
temp["datetime"] = pd.to_datetime(temp["date"]) + pd.to_timedelta(
    temp["hr"] - 1, unit="h"
)

# drop original columns
temp_final = temp.drop(columns=["date", "hr"])

# move datetime to first column
cols = ["datetime"] + [c for c in temp_final.columns if c != "datetime"]
temp_final = temp_final[cols].sort_values("datetime").reset_index(drop=True)
```

### Merging & Feature Engineering

```python
df = (
    load_final.merge(temp_final, on="datetime")
    .merge(temp_final, on="datetime")
    .merge(rh_final, on="datetime")
)
df.columns
```

```python
# date column breakdown into components
df["hour"] = df["datetime"].dt.hour
df["dayofweek"] = df["datetime"].dt.dayofweek
df["month"] = df["datetime"].dt.month
df["dayofyear"] = df["datetime"].dt.dayofyear
df["weekend"] = (df["dayofweek"] >= 5).astype(int)
```

```python
import numpy as np

# cyclic encoding (sin, cos) for time features
df["hour_sin"] = np.sin(2 * np.pi * df["hour"] / 24)
df["hour_cos"] = np.cos(2 * np.pi * df["hour"] / 24)

df["dow_sin"] = np.sin(2 * np.pi * df["dayofweek"] / 7)
df["dow_cos"] = np.cos(2 * np.pi * df["dayofweek"] / 7)

df["doy_sin"] = np.sin(2 * np.pi * df["dayofyear"] / 365)
df["doy_cos"] = np.cos(2 * np.pi * df["dayofyear"] / 365)
```

```python
# lag features: previous hour, previous day, previous week
df["hours_lag"] = df["demand"].shift(1)
df["day_lag"] = df["demand"].shift(24)
df["week_lag"] = df["demand"].shift(168)
previous_year_lag = 365 * 24
df["year_lag"] = df["demand"].shift(previous_year_lag)
```

```python
# rolling means based on past data only
df["rollmean_24"] = df["demand"].shift(1).rolling(24).mean()
df["rollmean_168"] = df["demand"].shift(1).rolling(168).mean()

# rolling standard deviation can also help
df["rollstd_24"] = df["demand"].shift(1).rolling(24).std()
```

```python
# Fourier terms for daily, weekly, yearly seasonality
df["t"] = np.arange(len(df))

K = 3  # number of harmonics

for k in range(1, K + 1):
    df[f"daily_sin_{k}"] = np.sin(2 * np.pi * k * df["t"] / 24)
    df[f"daily_cos_{k}"] = np.cos(2 * np.pi * k * df["t"] / 24)

    df[f"weekly_sin_{k}"] = np.sin(2 * np.pi * k * df["t"] / 168)
    df[f"weekly_cos_{k}"] = np.cos(2 * np.pi * k * df["t"] / 168)

    df[f"yearly_sin_{k}"] = np.sin(2 * np.pi * k * df["t"] / 8760)
    df[f"yearly_cos_{k}"] = np.cos(2 * np.pi * k * df["t"] / 8760)
```

```python
df = df.dropna(subset=["datetime"])
```

```python
# refactorize
def feature_engineering(df):

    # Date/time features
    df["hour"] = df["datetime"].dt.hour
    df["dayofweek"] = df["datetime"].dt.dayofweek
    df["month"] = df["datetime"].dt.month
    df["dayofyear"] = df["datetime"].dt.dayofyear
    df["weekend"] = (df["dayofweek"] >= 5).astype(int)

    # Cyclic encoding
    df["hour_sin"] = np.sin(2 * np.pi * df["hour"] / 24)
    df["hour_cos"] = np.cos(2 * np.pi * df["hour"] / 24)
    df["dow_sin"] = np.sin(2 * np.pi * df["dayofweek"] / 7)
    df["dow_cos"] = np.cos(2 * np.pi * df["dayofweek"] / 7)
    df["doy_sin"] = np.sin(2 * np.pi * df["dayofyear"] / 365)
    df["doy_cos"] = np.cos(2 * np.pi * df["dayofyear"] / 365)

    # Lag features
    df["hours_lag"] = df["demand"].shift(1)
    df["day_lag"] = df["demand"].shift(24)
    df["week_lag"] = df["demand"].shift(168)
    previous_year_lag = 365 * 24
    df["year_lag"] = df["demand"].shift(previous_year_lag)

    # Rolling statistics
    df["rollmean_24"] = df["demand"].shift(1).rolling(24).mean()
    df["rollmean_168"] = df["demand"].shift(1).rolling(168).mean()
    df["rollstd_24"] = df["demand"].shift(1).rolling(24).std()

    # Fourier terms
    df["t"] = np.arange(len(df))
    K = 3  # number of harmonics
    for k in range(1, K + 1):
        df[f"daily_sin_{k}"] = np.sin(2 * np.pi * k * df["t"] / 24)
        df[f"daily_cos_{k}"] = np.cos(2 * np.pi * k * df["t"] / 24)
        df[f"weekly_sin_{k}"] = np.sin(2 * np.pi * k * df["t"] / 168)
        df[f"weekly_cos_{k}"] = np.cos(2 * np.pi * k * df["t"] / 168)
        df[f"yearly_sin_{k}"] = np.sin(2 * np.pi * k * df["t"] / 8760)
        df[f"yearly_cos_{k}"] = np.cos(2 * np.pi * k * df["t"] / 8760)

    # Drop rows with missing datetime
    df = df.dropna(subset=["datetime"])

    return df
```

```python
load_fe = feature_engineering(load_final)
load_fe = load_fe.dropna()
```

```python
# split train/test and separate features/target
def split_train_test(df):
    df = df.sort_values("datetime")

    # test set = last year
    test_start = df["datetime"].max() - pd.DateOffset(years=1)

    train = df[df["datetime"] < test_start]
    test = df[df["datetime"] >= test_start]

    X_train = train.drop(columns=["demand", "datetime"])
    y_train = train["demand"]

    X_test = test.drop(columns=["demand", "datetime"])
    y_test = test["demand"]

    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = split_train_test(load_fe)
```

## Gradient Boost Tree Model

```python
from xgboost import XGBRegressor
from sklearn.model_selection import TimeSeriesSplit, RandomizedSearchCV
from sklearn.metrics import mean_absolute_percentage_error

tscv = TimeSeriesSplit(n_splits=5)

xgb = XGBRegressor(objective="reg:squarederror", random_state=42, n_jobs=-1)
```

```python
# hyperparameter grid for tuning
param_grid = {
    "n_estimators": [300, 500, 800],
    "max_depth": [4, 6, 8],
    "learning_rate": [0.01, 0.05, 0.1],
    "subsample": [0.7, 0.8, 1.0],
    "colsample_bytree": [0.7, 0.8, 1.0],
}

search = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=param_grid,
    n_iter=20,
    cv=tscv,
    scoring="neg_mean_absolute_percentage_error",
    n_jobs=-1,
    verbose=1,
)

search.fit(X_train, y_train)
```

```python
best_model_xgb = search.best_estimator_
```

## Export

```python
def export_submission(forecast_values, filename):
    # extract forecast values
    fc_values = pd.Series(forecast_values).astype(float).values

    # create July 2011 dates
    dates = pd.date_range(start="2011-07-01", end="2011-07-31", freq="D")

    # keep only first 31 forecast values
    df = pd.DataFrame({"date": dates, "load": fc_values[: len(dates)]})

    # write csv
    df.to_csv(filename, index=False)
```
